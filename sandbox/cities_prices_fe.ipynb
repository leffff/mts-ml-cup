```python
def get_site_name(row):  
    site=str(row).split('.')
    site[0]=site[0][7:]
    if(len(site)>=2):
        site='http://'+site[-2]+'.'+site[-1]
    else:
        site='NULL'
    return site

df = pandas_reduce_mem_usage( pd.read_csv('sites_out_copy.csv', sep=';', index_col=0))
df['site_url']=df['url_host'].apply(lambda x: get_site_name(x))
df['vpn_only']=df['text'].str.contains('403 Forbidden | Akado').fillna(False).astype(int)
df = df[df['site_url'].str.contains('userapi')==0]
df = df[df['vpn_only']==1].reset_index(drop= True )
df.to_csv('sites_out_vpn_only.csv')
```

```python
sourceFileName = 'sites_out_copy.xls'
targetFileName = 'sites_out_utf_copy.xls'
```

```python
import codecs
BLOCKSIZE = 1048576 # or some other, desired size in bytes
with codecs.open(sourceFileName, "r", "mac-roman") as sourceFile:
    with codecs.open(targetFileName, "w", "utf-8") as targetFile:
        while True:
            contents = sourceFile.read(BLOCKSIZE)
            if not contents:
                break
            targetFile.write(contents)
```
