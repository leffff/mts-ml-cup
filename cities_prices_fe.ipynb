```python
def long_lat(city_name, cities_df):
    long = cities_df['Широта'][cities_df['Название города']==city_name]
    lat = cities_df['Долгота'][cities_df['Название города']==city_name]
    return long,lat
```

```python
tqdm.pandas()
df[['long','lat']]=df['city_name'].progress_apply(lambda x: long_lat(x, cities))
```
